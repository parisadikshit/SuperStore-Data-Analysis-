# 3. Find out the per unit price from the data --A

```python
# converting date time field into pandas object
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

Cost Per Item Purchased = Selling Price per Item - Profit per Item

```python
# cost price = selling price - profit
data['Price per unit'] = ((data['Sales']/data['Quantity']))-((data['Profit']/data['Quantity'])) 
# Column 'Price per unit' is added at the end
```

# 4.Find out the monthly revenue and analyze the findings

```python
data.set_index(data['Ship Date'])
```

```python
data['Month']=0
data['Year']=0
# initially the data 
```

```python
# 2 new columns were created 
for i in range(len(data['Row ID'])):
    data['Month'][i]=data['Ship Date'][i].month
    data['Year'][i]=data['Ship Date'][i].year
```

```python
data.head()
```

```python
data_yearly = data.groupby('Year')
```

```python
data_yearly_year_2014 = data_yearly.get_group(2014)
data_yearly_year_2014.shape
```

```python
data_yearly_year_2015 = data_yearly.get_group(2015)
data_yearly_year_2015.shape
```

```python
data_yearly_year_2016 = data_yearly.get_group(2016)
data_yearly_year_2016.shape
```

```python
data_yearly_year_2017 = data_yearly.get_group(2017)
data_yearly_year_2017.shape
```

```python
data_yearly_year_2018 = data_yearly.get_group(2018)
data_yearly_year_2018.shape
```

```python
sales_month  = data[['Month','Sales','Year']]
sales_month.set_index(['Month','Year'])
```

```python
month=[]
yearss=[]
salesmonthyear=[]
data_month = sales_month.groupby(['Month','Year'])
for m,df in data_month:
    month.append(m[0])
    yearss.append(m[1])
    salesmonthyear.append(df['Sales'].sum())
```

```python
salemonthyeardf = pd.DataFrame({'Month':month,'Year':yearss,'Sales':salesmonthyear})
salemonthyeardf.head()
```

```python

sns.scatterplot(x=salemonthyeardf['Month'],y=salemonthyeardf['Sales'],hue=salemonthyeardf['Year'])
```

```python
sns.barplot(x=salemonthyeardf['Month'],y=salemonthyeardf['Sales'],hue=salemonthyeardf['Year'])
```

After plotting the monthly data(considering the year), In 2014, in **September**, the revenue collected is highest and the least is in february. In 2015, the revenue in the month of **December** is highest while in february is the lowest.In 2016, the revenue collected in the month of **September** is highest,in 2017, the revenue is highest in **December** and in 2018, the revenue is highest in **January**

''**In every year, in month of FEBRUARY the revenue seems to drop in comparison to other months**''

#  5.Find out the yearly revenue and analyze the findings

```python
# now grouping data by year
sales_year  = data[['Year','Sales']]
sales_year.set_index('Year')
```

```python
year = []
salesyear=[]
data_yearly = data.groupby('Year')
for y,df in data_yearly:
    year.append(y)
    salesyear.append(df['Sales'].sum())# so there is info of 5 years
```

```python
sales_year_df = pd.DataFrame({'Year':year,'Sales':salesyear})
sales_year_df.head()
```

```python
plt.bar(sales_year_df['Year'],sales_year_df['Sales'])
plt.xlabel('Years')
plt.ylabel('Sales')
plt.show('Yearly Revenue')
plt.show()
```

**From yearly study of revenue, it can be seen that the sales grew higher from 2014 till 2017 andthe sales were highest in 2017 and then suddenly dropped in 2018**
